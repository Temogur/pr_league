```python
import sqlite3


conn = sqlite3.connect('pr_league.sqlite3')
cursor = conn.cursor()
```

```python

cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()

for table in tables:
    print(table)
```

```python

cursor.execute("SELECT COUNT (*) FROM player WHERE Goals > 5")
results = cursor.fetchone()[0]
print(results)
#  პრინტავს იმ ფეხბურთელების რაოდენობას ბაზიდან რომლებსაც 5-ზე მეტი გოლი აქვთ გატანილი
```

```python

name = input("შეიყვანეთ მოთამაშის სახელი: ")
goals = int(input("შეიყვანეთ მოთამაშის გატანილი გოლების რაოდენობა: "))
position = input("შეიყვანეთ მოთამაშის პოზიცია: ")

cursor.execute("INSERT INTO player (PlayerName, Goals , Position) VALUES (?, ?, ?)", (name, goals, position))
conn.commit()
# ამით დავამატებთ ახალი მოთამაშეს
```

```python

new_goals = int(input("შეიყვანეთ მოთამაშის გატანილი გოლების რაოდენობა: "))
Player = input("შეიყვანეთ მოთამაშის სახელი: ")

cursor.execute("UPDATE player SET Goals = ? WHERE PlayerName = ?", (new_goals, Player))
conn.commit()
# მოთამაშის  სტატისტიკა განვაახლე 
```

```python
Player = input("შეიყვანეთ წასაშლელი მოთამაშის სახელი: ")

cursor.execute("DELETE FROM player WHERE PlayerName = ?", (Player,))
conn.commit()
# მოთამაშის წაშლა სახელის მიხედვით
```

```python
import matplotlib.pyplot as plt

# გოლების მიხედვით მოთამაშეების რაოდენობა
cursor.execute("SELECT Goals, COUNT(*) FROM player GROUP BY Goals")
data = cursor.fetchall()

Goals = [row[0] for row in data]
counts = [row[1] for row in data]

#  სვეტოვანი დიაგრამა
plt.bar(Goals, counts)
plt.title('მოთამაშეების რაოდენობა გოლების მიხედვით')
plt.xlabel('გოლი')
plt.ylabel('რაოდენობა')
plt.show()
```

```python
plt.pie(counts, labels=Goals, autopct='%1.1f%%')
plt.title('გოლების წილი მოთამაშეებს შორის')
plt.show()
# წრიული დიაგრამა
```

```python
plt.plot(Goals, counts, marker='o')
plt.title('გოლების განაწილება მოთამაშეებში')
plt.xlabel('გოლი')
plt.ylabel('რაოდენობა')
plt.show()
# ხაზოვანი დიაგრამა
```
